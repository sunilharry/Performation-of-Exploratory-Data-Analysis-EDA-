```python
import pandas as pd

data = pd.read_csv("C:/Users/ssuun/Downloads/data.xlsx - Sheet1.csv")
print(data.head())
print("Shape of dataset:", data.shape)
print(data.describe())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting histograms for numerical columns
numerical_columns = ['Salary', '10percentage', '12percentage', 'collegeGPA', 'English', 'Logical', 'Quant']

for col in numerical_columns:
    plt.figure(figsize=(8,5))
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()

# Boxplots to find outliers
for col in numerical_columns:
    plt.figure(figsize=(8,5))
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col}')
    plt.show()

# Countplot for a categorical column
plt.figure(figsize=(8,5))
sns.countplot(data['Gender'])
plt.title('Frequency Distribution of Gender')
plt.show()
```

```python
# Scatter plot for numerical variables
sns.scatterplot(x='collegeGPA', y='Salary', data=data)
plt.title('Scatter plot between CollegeGPA and Salary')
plt.show()

# Box plot for categorical and numerical variables
sns.boxplot(x='Gender', y='Salary', data=data)
plt.title('Box plot between Gender and Salary')
plt.show()

# Stacked bar plot for categorical-categorical relationship
pd.crosstab(data['Gender'], data['Specialization']).plot(kind='bar', stacked=True)
plt.title('Stacked Bar Plot of Gender vs Specialization')
plt.show()
```

```python

from scipy.stats import chi2_contingency
contingency_table = pd.crosstab(data['Gender'], data['Specialization'])
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f'Chi-square Statistic: {chi2}, p-value: {p}')
```
